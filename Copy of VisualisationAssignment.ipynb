```python
# 4. Generate a dataset with categories and numerical values. Visualize the distribution of a numerical variable across different categories


np.random.seed(42)


categories = ['A', 'B', 'C', 'D']
data = {
    'Category': np.random.choice(categories, size=100),  # Randomly assign categories
    'Value': np.random.randn(100) * 10 + 50  # Normally distributed numerical values
}


df = pd.DataFrame(data)


print(df.head())

import matplotlib.pyplot as plt

# Bar chart
plt.figure(figsize=(10, 6))
df.groupby('Category')['Value'].mean().plot(kind='bar', color='skyblue')
plt.title('Average Value by Category')
plt.xlabel('Category')
plt.ylabel('Average Value')
plt.grid(axis='y')
plt.show()



# Box plot
plt.figure(figsize=(10, 6))
sns.boxplot(x='Category', y='Value', data=df, palette='Set2')
plt.title('Distribution of Value by Category')
plt.xlabel('Category')
plt.ylabel('Value')
plt.grid(True)
plt.show()

```

```python
#5. Generate a synthetic dataset with correlated features. Visualize the correlation matrix of a dataset using a heatmap.


np.random.seed(42)


n_samples = 100

X = np.random.randn(n_samples, 3)


X[:, 1] = X[:, 0] * 0.5 + np.random.randn(n_samples) * 0.5  # Feature 1 correlated with Feature 0
X[:, 2] = X[:, 0] * 0.2 + X[:, 1] * 0.8 + np.random.randn(n_samples) * 0.5  # Feature 2 correlated with Feature 0 and Feature 1

df = pd.DataFrame(X, columns=['Feature 0', 'Feature 1', 'Feature 2'])


print(df.head())


corr_matrix = df.corr()
print(corr_matrix)



plt.figure(figsize=(10, 8))

sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)


plt.title('Correlation Matrix Heatmap')
plt.show()

```

```python
#4 . Using the given x and y data, generate a 3D surface plot to visualize the function

x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
x, y = np.meshgrid(x, y)
z = np.sin(np.sqrt(x**2 + y**2))
data = {
 'X': x.flatten(),
 'Y': y.flatten(),
 'Z': z.flatten()
 }

df = pd.DataFrame(data)

df

import plotly.graph_objects as go


X = df['X'].values.reshape(100, 100)
Y = df['Y'].values.reshape(100, 100)
Z = df['Z'].values.reshape(100, 100)

fig = go.Figure(data=[go.Surface(z=Z, x=X, y=Y)])


fig.update_layout(title='3D Surface Plot of sin(sqrt(x^2 + y^2))',
                  scene=dict(xaxis_title='X', yaxis_title='Y', zaxis_title='Z'))


fig.show()
```

```python
#5 Using the given dataset, create a bubble chart to represent each country's population (y-axis), GDP (xaxis), and bubble size proportional to the population.





np.random.seed(25)


data = {
    'Country': ['USA', 'Canada', 'UK', 'Germany', 'France'],
    'Population': np.random.randint(100, 1000, 5),
    'GDP': np.random.randint(500, 2000, 5)
}


df = pd.DataFrame(data)


plt.figure(figsize=(10, 6))
plt.scatter(df['GDP'], df['Population'],
            s=df['Population'],
            alpha=0.5,
            c='blue',
            edgecolors='w',
            linewidth=2)


plt.title('Bubble Chart of Countries by GDP and Population')
plt.xlabel('GDP')
plt.ylabel('Population')


for i, country in enumerate(df['Country']):
    plt.annotate(country, (df['GDP'][i], df['Population'][i]),
                 fontsize=12, ha='right')

plt.grid(True)
plt.show()


```

```python
##  Bokeh Assignment ##

#1 Create a Bokeh plot displaying a sine wave. Set x-values from 0 to 10 and y-values as the sine of x.

from bokeh.plotting import figure, show
from bokeh.io import output_notebook


output_notebook()


x = np.linspace(0, 10, 100)
y = np.sin(x)

p = figure(title="Sine Wave", x_axis_label='x', y_axis_label='y')


p.line(x, y, legend_label="y=sin(x)", line_width=2)


show(p)


```

```python
#2 .Create a Bokeh scatter plot using randomly generated x and y values. Use different sizes and colors for the markers based on the 'sizes' and 'colors' columns.



output_notebook()

np.random.seed(42)
data = {
    'x': np.random.uniform(0, 10, 100),
    'y': np.random.uniform(0, 10, 100),
    'sizes': np.random.uniform(10, 50, 100),
}

df = pd.DataFrame(data)


p = figure(title="Random Scatter Plot with Variable Sizes and Colors", x_axis_label='x', y_axis_label='y')


p.circle(x='x', y='y', size='sizes', color='colors', fill_alpha=0.6, source=df, legend_label="Random Points")


show(p)
```

```python
#3 Generate a Bokeh bar chart representing the counts of different fruits using the following dataset.



from bokeh.transform import factor_cmap
from bokeh.models import ColumnDataSource


output_notebook()


fruits = ['Apples', 'Oranges', 'Bananas', 'Pears']
counts = [20, 25, 30, 35]


source = ColumnDataSource(data=dict(fruits=fruits, counts=counts))

p = figure(x_range=fruits, title="Fruit Counts", x_axis_label='Fruit', y_axis_label='Count')


p.vbar(x='fruits', top='counts', width=0.7, source=source,
       legend_field="fruits", line_color='white', fill_color=factor_cmap('fruits', palette='Spectral4', factors=fruits))

p.legend.title = "Fruits"
p.xgrid.grid_line_color = None
p.y_range.start = 0


show(p)
```
